```python
import numpy as np
```

```python
class Neuron:

    def __init__ (self, input, weight):
      self.input = input
      self.weight = weight

    def summation(self):
      n = np.dot(self.input, self.weight)
      print (n)

    def relu(self, x):
        return (x > 0) * x
```

image.png

```python
#Weights and inputs

W_1 = 1,3,5,7,9
W_2 = 5,4,3,2,1
x_in = 5,6,7,8,9

#taking of value
S1 = Neuron(x_in, W_1)
S2 = Neuron(x_in, W_1)
S3 = Neuron(x_in, W_1)
S4 = Neuron(x_in, W_1)
S5 = Neuron(x_in, W_1)
```

```python
#Summation of all values

S1.summation()
S2.summation()
S3.summation()
S4.summation()
S5.summation()
```

```python
#Getting the value of all outputs
Output = S1.relu(195), S2.relu(195), S3.relu(195), S4.relu(195), S5.relu(195)
```

```python
#Second Neuron
S2 = Neuron(Output,W_2)
```

```python
#Summation of Second Neuron
S2.summation()
S2
```
